```python
#-----------------------------#
#     Hypsometric Curve       #
#-----------------------------#

class hypsometry():
    
    def __init__(self, DEM):
        self.DEM = DEM
    
    def get_hypcurve(self):
        DEM_banded = self.DEM.ravel()
        DEM_norm = DEM_banded / np.max(self.DEM)
        DEM_norm[DEM_norm < 0.0000001] = -9999
        h = np.histogram(DEM_norm, bins=20, range=(0, 1))
        l = ((((np.cumsum(h[0]))*(50**2))/(1000**2)),h[1])
        from bokeh.charts import Bar, Area, show
        j = figure(title = "Hypsometric Curve", plot_width=900, plot_height=600)
        j.title.text_font_size = "25px"
        j.yaxis.axis_label = 'Normalized Cumulative Elevation'
        j.xaxis.axis_label = 'Area (km2)'
        lx = l[0]
        ly = [ 0.0, 0.0625, 0.125, 0.1875, 0.25, 0.3125, 0.375, 0.4375, 0.5, 
              0.5625, 0.625, 0.6875, 0.75, 0.8125, 0.875, 0.9375, 1.0 ]
        j.line(lx[3:], ly, line_width=2)
        j.circle(lx[3:], ly, fill_color="white", size=8)
        output_notebook()
        return show(j)


dh = hypsometry(DEM)
dh.get_hypcurve()
```

```python
#-----------------------------------------------------------#
#  Create SWE Variability Mask Method and Snowline Method   #
#-----------------------------------------------------------#


class SWE_arrays():
    
    """
    This function with output with defined swe_map (from array list), dem_array, aspect_map, and veg_map
   
    #-----------------------------#
    #      create_mask()          #
    #-----------------------------#
    User-defined inputs are:
    2 value lists with the desired range in elevation and aspect
    and 0 for forested or 1 for unforested regions
        
        for example: create_mask(swe_1, DEM, aspect, veg, [0,180], [1800,2500], 0)
        #    for all northern slopes between east and west 
        #    and elevation band between 1800 and 2500 meters
        #    and forested region
    """
    
    ## range elevation = 910.07727 to 3965.31010
    ## range aspect = 0° means East, 90° North, 180° West, 270° South
    ## pick vegetation = 0 (non-forested) or 1 (forested)
    
    def __init__(self, swe_array, DEM, aspect, veg_map, name):
        self.swe_array = swe_array
        self.DEM = DEM
        self.aspect = aspect
        self.name = name
        self.veg_map = veg_map
   
    def getname(self):
        return self.name

    def create_mask(self, direction_range, elevation_range, veg_value):
        
        """Creates mask based on aspect, elevation and forest cover. 
        First LIST is trigonometric aspect
        Second LIST is elevation range
        Third VALUE is 0 to MASK forested and 1 to MASK unforested"""
        
        aspect_range = ma.masked_outside(self.aspect, np.min(direction_range), np.max(direction_range))
        DEM_range = ma.masked_outside(self.DEM, np.min(elevation_range), np.max(elevation_range))
        forested = ma.masked_not_equal(self.veg_map, veg_value)
        
        """Combines masks"""
        
        aspect_mask = ma.getmask(aspect_range)
        DEM_mask = ma.getmask(DEM_range)
        mask_subtotal = ma.mask_or(aspect_mask, DEM_mask)
        
        forest_mask = ma.getmask(forested)
        mask_total = ma.mask_or(forest_mask, mask_subtotal)
        
        """Applies to swe array and outputs"""
        
        swe_mask = ma.array(self.swe_array, mask = mask_total)
        
        out_block = swe_mask
        return out_block
        
    def snowline(self):
        
        """Finds the approximate snowline at given timestep by returning the average elevation """
        snow_1 = ma.masked_less(self.swe_array, 0.02)
        snow_2 = ma.masked_greater(self.swe_array, 0.05)
        
        snow_mask1 = ma.getmask(snow_1)
        snow_mask2 = ma.getmask(snow_2)
        snow_total = ma.mask_or(snow_mask1, snow_mask2)
        
        DEM_nonzero = ma.masked_equal(self.DEM, 0)
        DEM_mask1 = ma.getmask(DEM_nonzero)
        DEM_total = ma.mask_or(DEM_mask1, snow_total)
        
        DEM_new = ma.array(DEM, mask= DEM_total)
        data = DEM_new.mean()
        return data   
```

```python
#------------------------#
#  Create Facet Heatmap  #
#------------------------#

### Create keys to select rows in function
NF = "North Forested"
SF = "South Forested"
NuF = "North Unforested"
SuF = "South Unforested"

def facet_heatmap(_NF, _NuF, _SF, _SuF, Title, cbm):
    vmax = cbm
    cbarkws = {"fraction": 0.1}
    
    fig = plt.figure(1, figsize=(9,6))
    plt.suptitle(Title, fontsize=28)
    ax1 = plt.subplot(221)
    ax1.set_title("North Forested")
    sns.heatmap(pd.DataFrame.from_csv(_NF), vmax=vmax, xticklabels=False, cbar_kws=cbarkws)
    ax2 = plt.subplot(222)
    ax2.set_title("North Unforested")
    sns.heatmap(pd.DataFrame.from_csv(_NuF), vmax= vmax, xticklabels=False, yticklabels=False, cbar_kws=cbarkws)
    ax3 = plt.subplot(223)
    ax3.set_title("South Forested")
    sns.heatmap(pd.DataFrame.from_csv(_SF), vmax=vmax, cbar_kws=cbarkws)
    ax4 = plt.subplot(224)
    ax4.set_title("South Unforested")
    sns.heatmap(pd.DataFrame.from_csv(_SuF), vmax=vmax, yticklabels=False, cbar_kws=cbarkws)
    #fig.tight_layout()
    plt.show

#------------------------#
#    Plot Facet Heatmap  #
#------------------------#    

facet_heatmap(NF_14, NuF_14, SF_14, SuF_14, "2014", 0.4)

#plt.savefig("2016.png", dpi=600)
```

```python
#----------------------------#
#  Export raster from array  #
#----------------------------#

"""This function is here to export to ArcMap readable file"""

def array_to_raster(array):
    """Array > Raster
    Save a raster from a C order array.

    :param array: ndarray
    """
    dst_filename = 'aspect_output.tiff'
    
    
    pjtn = gdal.Open('Tuol_dem.tif')

    # You need to get those values from data.GetGeoTransform()
    x_pixels = 1062  # number of pixels in x
    y_pixels = 1007  # number of pixels in y
    PIXEL_SIZE = 50.0  # size of the pixel...        
    x_min = 254247.0  
    y_max = 4229667.0  # x_min & y_max are like the "top left" corner.
    wkt_projection = pjtn.GetProjection()

    driver = gdal.GetDriverByName('GTiff')

    dataset = driver.Create(
        dst_filename,
        x_pixels,
        y_pixels,
        1,
        gdal.GDT_Float32, )

    dataset.SetGeoTransform((
        x_min,    # 0
        PIXEL_SIZE,  # 1
        0,                      # 2
        y_max,    # 3
        0,                      # 4
        -PIXEL_SIZE))  

    dataset.SetProjection(wkt_projection)
    dataset.GetRasterBand(1).WriteArray(array)
    dataset.FlushCache()  # Write to disk.
    return dataset, dataset.GetRasterBand(1) 
```
